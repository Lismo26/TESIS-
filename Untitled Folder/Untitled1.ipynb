```python
# Define una función para aplicar la lógica condicional
def calcular_puntaje(row):
    if row['CONDICION'] == 'NO SELECCIONADO':
        return int(row['PUNTAJEFINAL']) - row['ALEA']
    else:
        return int(row['PUNTAJEFINAL']) + row['ALEA']

# Aplica la función a cada fila utilizando apply()
data2['Puntaje_RD'] = data2.apply(calcular_puntaje, axis=1)
```
