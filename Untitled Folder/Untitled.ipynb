```python
import numpy as np
import pandas as pd
import random

# Crear la columna de aleatorización
data2['ALEA'] = [random.randint(0, 500) / 100 for _ in range(len(data2))]

# Definir la función para calcular el valor transformado según las condiciones dadas
def transform_value(row):
    if row['PUNTAJEFINAL'] == 60:
        if row['CONDICION'] == 'NO SELECCIONADO':
            return int(row['PUNTAJEFINAL']) - row['ALEA']
        elif row['CONDICION'] == 'SELECCIONADO':
            return int(row['PUNTAJEFINAL']) + row['ALEA']
    else:
        # Cuando el valor de 'PUNTAJEFINAL' es diferente de 60
        return int(row['PUNTAJEFINAL']) + row['ALEA'] if random.choice([True, False]) else int(row['PUNTAJEFINAL']) - row['ALEA']
# Aplicar la transformación a la columna 'PUNTAJEFINAL'
data2['PUNTAJEFINAL_TRANSFORMADO'] = data2.apply(transform_value, axis=1)

# Definir los datos para la regresión
y = data2['matriculado']
score = data2['PUNTAJEFINAL_TRANSFORMADO']

# Realizar bootstrap para obtener los coeficientes beta
num_bootstrap = 100
beta_samples = []

for i in range(num_bootstrap):
    # Generar una muestra bootstrap aleatoria con reemplazo
    indices_bootstrap = np.random.choice(len(y), size=len(y), replace=True)
    y_bootstrap = y.iloc[indices_bootstrap] if isinstance(y, pd.Series) else y[indices_bootstrap]
    x_bootstrap = score.iloc[indices_bootstrap] if isinstance(score, pd.Series) else score[indices_bootstrap]
    
    # Realizar la regresión en la muestra bootstrap
    
    covars = pd.get_dummies( data2.IES ).astype(int).iloc[:, 1:]
    est_bootstrap = rdrobust(y_bootstrap, x_bootstrap, c=60, covs = covars, all = True)  # Asumiendo que rdrobust es la función que realiza la regresión discontinua
    # Almacenar los coeficientes beta
beta_samples.append(est_bootstrap.coef.values.flatten())

    
# Convertir la lista de betas en un DataFrame de pandas
df_betas = pd.DataFrame(beta_samples, columns=['Beta_' + str(i) for i in range(1, len(beta_samples[0]) + 1)])

# Guardar el DataFrame en un archivo CSV
df_betas.to_csv('betas_bootstrap.csv', index=False)
```
